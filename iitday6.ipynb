```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."

# Example usage
print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # Error message
```

```python
# Create a sample log file
with open("system.log", "w") as f:
    f.write("INFO: System started\n")
    f.write("ERROR: Disk full\n")
    f.write("ERROR: Network timeout\n")
    f.write("ERROR: Disk full\n")

def analyze_log(filename):
    error_counts = {}
    with open(filename, "r") as f:
        for line in f:
            if "ERROR" in line:
                parts = line.strip().split("ERROR:")
                if len(parts) > 1:
                    error_type = parts[1].strip()
                    error_counts[error_type] = error_counts.get(error_type, 0) + 1
    return error_counts

# Run it
errors = analyze_log("system.log")
print("Error counts:", errors)
```

```python
import csv

# Create a sample grade sheet
with open("grades.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Math", "Science", "English"])
    writer.writerow(["Alice", 85, 90, 88])
    writer.writerow(["Bob", 70, 75, 80])
    writer.writerow(["Charlie", 95, 92, 96])

def process_grades(input_file, output_file):
    results = [["Name", "Average"]]
    with open(input_file, "r") as f:
        reader = csv.reader(f)
        header = next(reader)  # skip header
        for row in reader:
            name = row[0]
            marks = list(map(float, row[1:]))
            avg = sum(marks) / len(marks)
            results.append([name, avg])
    
    with open(output_file, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerows(results)

# Run it
process_grades("grades.csv", "averages.csv")

# Show results
with open("averages.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# Create a sample text file
with open("sample.txt", "w") as f:
    f.write("Python is fun.\n")
    f.write("I love coding in Python.\n")
    f.write("Data science uses Python a lot.\n")
    f.write("Java is also popular.\n")

def search_keyword(filename, keyword):
    matches = []
    with open(filename, "r") as f:
        for i, line in enumerate(f, 1):
            if keyword.lower() in line.lower():
                matches.append((i, line.strip()))
    return matches

# Run it
results = search_keyword("sample.txt", "Python")
for line_no, text in results:
    print(f"Line {line_no}: {text}")
```
