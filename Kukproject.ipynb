```python
df["Income"] = df["Income"].fillna(df["Income"].median())
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
from sklearn.preprocessing import StandardScaler

features = df_encoded.select_dtypes(include=np.number).drop(columns=["ID"], errors="ignore")
scaler_std = StandardScaler()
X_std = scaler_std.fit_transform(features)
```

```python
from scipy import stats

# Drop constant columns (std = 0) before z-score calc
constant_cols = features.columns[features.std() == 0]
print("Constant columns found:", list(constant_cols))

features_clean = features.drop(columns=constant_cols)

z_scores = np.abs(stats.zscore(features_clean))
df_no_outliers = df_encoded[(z_scores < 3).all(axis=1)]

print("Before:", df_encoded.shape, "After:", df_no_outliers.shape)
```

```python
# Convert Dt_Customer to datetime
df_no_outliers["Dt_Customer"] = pd.to_datetime(df_no_outliers["Dt_Customer"], format="%d-%m-%Y")

# Create a numeric feature: days since customer enrolled (relative to most recent enrollment date)
latest_date = df_no_outliers["Dt_Customer"].max()
df_no_outliers["Customer_Tenure_Days"] = (latest_date - df_no_outliers["Dt_Customer"]).dt.days

# Now drop the original date column since it's no longer needed
df_no_outliers = df_no_outliers.drop(columns=["Dt_Customer"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Use Income to predict MntWines
X_gd = df_no_outliers[["Income"]].values.astype(float)
y_gd = df_no_outliers["MntWines"].values.astype(float)

# Standardize (important for gradient descent to converge properly)
X_gd = (X_gd - X_gd.mean()) / X_gd.std()
y_gd = (y_gd - y_gd.mean()) / y_gd.std()

print(X_gd.shape, y_gd.shape)
```

```python
def compute_cost(X, y, w, b):
    m = len(y)
    preds = X.flatten() * w + b
    return np.mean((preds - y) ** 2) / 2
```

```python
def batch_gd(X, y, lr=0.01, epochs=100):
    m = len(y)
    X = X.flatten()
    w, b = 0.0, 0.0
    costs = []
    for _ in range(epochs):
        preds = X * w + b
        dw = (1/m) * np.sum(X * (preds - y))
        db = (1/m) * np.sum(preds - y)
        w -= lr * dw
        b -= lr * db
        costs.append(compute_cost(X.reshape(-1,1), y, w, b))
    return w, b, costs

w_b, b_b, costs_b = batch_gd(X_gd, y_gd)
print("Batch GD final w, b:", w_b, b_b)
```

```python
def sgd(X, y, lr=0.01, epochs=100):
    m = len(y)
    X = X.flatten()
    w, b = 0.0, 0.0
    costs = []
    for _ in range(epochs):
        for i in range(m):
            xi, yi = X[i], y[i]
            pred = xi * w + b
            dw = xi * (pred - yi)
            db = (pred - yi)
            w -= lr * dw
            b -= lr * db
        costs.append(compute_cost(X.reshape(-1,1), y, w, b))
    return w, b, costs

w_s, b_s, costs_s = sgd(X_gd, y_gd)
print("SGD final w, b:", w_s, b_s)
```

```python
def mini_batch_gd(X, y, lr=0.01, epochs=100, batch_size=32):
    m = len(y)
    X = X.flatten()
    w, b = 0.0, 0.0
    costs = []
    for _ in range(epochs):
        indices = np.random.permutation(m)
        X_shuf, y_shuf = X[indices], y[indices]
        for start in range(0, m, batch_size):
            xb = X_shuf[start:start+batch_size]
            yb = y_shuf[start:start+batch_size]
            preds = xb * w + b
            dw = (1/len(yb)) * np.sum(xb * (preds - yb))
            db = (1/len(yb)) * np.sum(preds - yb)
            w -= lr * dw
            b -= lr * db
        costs.append(compute_cost(X.reshape(-1,1), y, w, b))
    return w, b, costs

w_m, b_m, costs_m = mini_batch_gd(X_gd, y_gd)
print("Mini-Batch GD final w, b:", w_m, b_m)
```

```python
plt.figure(figsize=(9,6))
plt.plot(costs_b, label="Batch GD")
plt.plot(costs_s, label="SGD")
plt.plot(costs_m, label="Mini-Batch GD")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Convergence Comparison: Batch vs SGD vs Mini-Batch")
plt.legend()
plt.show()
```

### Compare convergence speed
Look at the plot: note which curve drops fastest in the first ~10 epochs, and which one is smoothest vs. noisiest. Typically Batch GD is smooth but slower per-epoch progress can look choppy if lr is off; SGD is noisy but often reaches a low cost faster in wall-clock terms; Mini-Batch usually balances both.

```python
plt.figure(figsize=(9,6))
for lr in [0.01, 0.001, 0.0001]:
    _, _, c = batch_gd(X_gd, y_gd, lr=lr, epochs=100)
    plt.plot(c, label=f"lr={lr}")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Effect of Learning Rate on Convergence")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6))
for bs in [8, 32, 128]:
    _, _, c = mini_batch_gd(X_gd, y_gd, batch_size=bs, epochs=100)
    plt.plot(c, label=f"batch_size={bs}")
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Effect of Batch Size on Convergence")
plt.legend()
plt.show()
```
