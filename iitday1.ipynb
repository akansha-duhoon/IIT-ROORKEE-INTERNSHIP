```python
height=float(input("enter value of height in metres:"))
weight=float(input("enter value of weight in kg:"))
BMI=weight/(height*height)
if BMI<18.5:
     print("category=underweight")
elif BMI<25:
     print("category=normal")
else:
     print("category=overweight")
```
