```python
# Iterative approach
def factorial_iterative(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
def factorial_recursive(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n-1)
a= int(input("enter the number:"))
print(f"Iterative factorial of {a}: {factorial_iterative(a)}")
print(f"Recursive factorial of {a}: {factorial_recursive(a)}")
```
