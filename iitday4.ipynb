```python
foods = []
prices = []
total = 0

while True:
    food = input("Enter the food (q to quit): ")
    if food.lower() == "q":
        break
    else:
        price = float(input("Enter the price of food (rupees): "))
        foods.append(food)
        prices.append(price)

print("----- My Cart -----")
for food in foods:
    print(food)

for price in prices:
    total += price

print()
print(f"My total is: ₹{total}")
```
